```python
import pandas as pd

#Carregando o dataset
df = pd.read_csv("/Users/dev_lucas_arrais/Desktop/projetos-ml/01-titanic/data/train.csv")

#Visualizando primeiras linhas 
df.head()
```

```python
#Quais colunas tem valores nulos e quais sao os tipos de dados?
df.info()
```

```python
#Resumo estatistico -> media, desvio, minimo,maximo...
df.describe()
```

```python
#Quantos dados faltam por coluna ?
df.isnull().sum()
```

```python
#Quais categorias sao exclusivas por coluna ?
df.nunique()
```

```python
#Removendo a coluna "Cabin", muitos NaNs
df = df.drop(columns = ['Cabin'])
```

```python
#Preenchendo "Embarked" com a moda e Idade com a mediana
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
df['Age'].fillna(df['Age'].median(), inplace = True)
```

```python
#Extraindo os titulos dos nomes
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df['Title'].value_counts()
```

```python
# Agrupar títulos raros
rare_titles = ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']
df['Title'] = df['Title'].replace({
    'Mlle': 'Miss',
    'Ms': 'Miss',
    'Mme': 'Mrs',
})
df['Title'] = df['Title'].replace(rare_titles, 'Rare')

# Verificando o resultado
df['Title'].value_counts()
```

```python
#Mapeando sexo ( binario -> mais simples para o modelo)
df['Sex'] = df['Sex'].map({'male': 0, 'female':1})
```

```python
# One-Hot Encoding de Embarked e Title
df = pd.get_dummies(df, columns=['Embarked', 'Title'], drop_first=True)
```

```python
# Remover colunas que não servem para o modelo
df = df.drop(columns=['Name', 'Ticket'])
```

```python
#Colocando todo o dataset com valores inteiros
df = df.astype(int)
```

```python
#Salvando dataset
df.to_csv("/Users/dev_lucas_arrais/Desktop/projetos-ml/01-titanic/data", index=False)
```
